Dynamics := taup * dydt = -y + Kp*u

```python
#Imports
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
from scipy.integrate import odeint

Kp = 3.0
taup = 2.0

num = [Kp]
den = [taup,1]

sys = signal.TransferFunction(num,den)

t,y = signal.step(sys)

plt.figure(1)
plt.plot(t,y,"g--")
plt.show()
```

```python
#Imports
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from scipy import signal
from scipy.integrate import odeint

def model(y,t):
    u = 1
    dydt = (-y + Kp * u) / taup
    return dydt

t = np.linspace(0,14,100)
y = odeint(model, 0,t)

plt.plot(t,y,"b--")
```
