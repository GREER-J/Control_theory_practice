```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# function that returns dz/dt
def model(z,t):
    r = z[0]
    f = z[1]
    k1 = 2 
    k2 = 1
    dfdt = k2*r*f - k2*f
    drdt = k1 * r - r*f
    dzdt = [drdt,dfdt]
    return dzdt

# initial condition
r0 = 1
f0 = 1
z0 = [r0,f0]

# time points
t = np.linspace(0,1000)

# solve ODE
z = odeint(model,z0,t)

# plot results
plt.plot(t,z[:,0],'b-',label='Rabits')
plt.plot(t,z[:,1],'r-',label="foxes")
plt.ylabel('response')
plt.xlabel('time')
plt.legend(loc='best')
plt.show()
```
